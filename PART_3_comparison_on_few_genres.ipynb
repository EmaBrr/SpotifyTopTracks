```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
df = pd.read_csv('cleaned_spotifytoptracks.csv')
df.head(5)
```

#### How does the danceability score compare between Pop, Hip-Hop/Rap, Dance/Electronic, and Alternative/Indie genres?

```python
genre_filtered = df[df['genre'].isin(['Pop', 'Hip-Hop/Rap', 'Dance/Electronic', 'Alternative/Indie'])]
genre_danceability = genre_filtered.groupby('genre')['danceability'].mean()

genre_danceability.plot(kind='bar', figsize=(6, 4))
plt.title("Average Danceability Score by Genre")
plt.xlabel("Genre")
plt.ylabel("Average Danceability")
plt.show()
```

#### How does the loudness score compare between Pop, Hip-Hop/Rap, Dance/Electronic, and Alternative/Indie genres?

```python
genre_filtered = df[df['genre'].isin(['Pop', 'Hip-Hop/Rap', 'Dance/Electronic', 'Alternative/Indie'])]
genre_danceability = genre_filtered.groupby('genre')['loudness'].mean()

genre_danceability.plot(kind='bar', figsize=(6, 4))
plt.title("Average Loudness Score by Genre")
plt.xlabel("Genre")
plt.ylabel("Average Loudness")
plt.show()
```
